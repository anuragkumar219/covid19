# digit recognition

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
digits
```

```python
#digits.images show an array
digits.images[0]
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
#to determine within 0-4 we are using target
digits.target[:5]
```

```python
from sklearn.linear_model import LogisticRegression
model4=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
model4.fit(x_train,y_train)
```

```python
model4.predict(x_test)
```

```python
y_test
```

```python
model4.score(x_test,y_test)
```

```python
#to know which were wrong predicted by our model we use confusion matrix
y_predicted=model4.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
#to have better ui we use seaborn library
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

# decision tree

```python
import pandas as pd
df6=pd.read_csv('salaries.csv')
df6.head()
```

```python
x=df6.iloc[:,[0,1,2]]
y=df6.iloc[:,3]
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
x['company_n']=le_company.fit_transform(x['company'])
x['job_n']=le_job.fit_transform(x['job'])
x['degree_n']=le_degree.fit_transform(x['degree'])
x
```

```python
x.drop(['company','job','degree'],axis='columns',inplace=True)
x
```

```python
from sklearn import tree
model5=tree.DecisionTreeClassifier()
```

```python
model5.fit(x,y)
```

```python
model5.score(x,y)
```

```python
model5.predict([[2,1,1]])
```

```python
import pandas as pd
df7=pd.read_csv('titanic.csv')
df7.head()
```

```python
x=df7.iloc[:,[2,4,5,9]]
y=df7.iloc[:,1]
```

```python
x['sex_n']=le_sex.fit_transform(x['Sex'])
```

```python
x
```

```python
x.drop(['Sex'],axis='columns',inplace=True)
```

```python
x
```

```python
model6.fit(x_train,y_train)
```
